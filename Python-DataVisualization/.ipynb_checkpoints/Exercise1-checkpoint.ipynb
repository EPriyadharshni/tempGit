```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python

df_SPX = pd.read_csv('https://s3.eu-west-1.amazonaws.com/neueda.conygre.com/pydata/SPX.csv',
                     index_col='Date', parse_dates=True)


print(df_SPX.dtypes)

df_SPX.head()
```

```python
df_SPX['Price']=pd.to_numeric(df_SPX['Price'].str.replace(',',''))
```

```python
df_SPX['Open']=pd.to_numeric(df_SPX['Open'].str.replace(',',''))
```

```python
df_SPX['High']=pd.to_numeric(df_SPX['High'].str.replace(',',''))
```

```python
df_SPX['Low']=pd.to_numeric(df_SPX['Low'].str.replace(',',''))
```

```python
df_SPX['Change %']=pd.to_numeric(df_SPX['Change %'].str.replace('%',''))
print(df_SPX.dtypes)

df_SPX.head()
```

```python
df_SPX.sort_values('Date')
```

```python
by_BMonth=pd.Grouper(freq='BQ')

fncts=['min','max','mean']
df_SPX.groupby(by_BMonth).agg(fncts)
```

```python
cols=['High','Low']
rows='Date>="01/06/2015" and Date <="30/06/2016"'
df=df_SPX.sort_values('Date').query(rows)[cols]
display(df)
```

```python
df.plot(kind='line',color=['r','b'],title='High and Low',figsize=(18,9))
```
